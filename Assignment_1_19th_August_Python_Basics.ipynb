**Ques 1.** Declare two variables, **'x'** and **'y'**, and assign them integer values. Swap the values of these variables without using any temporary variable.

**Ans 1.**

```python
# Declare and initialize variables
x = 5
y = 10

# Before swapping
print("Before swapping:")
print("x =", x)
print("y =", y)

# Swapping without using any temporary variable
x = x + y
y = x - y
x = x - y

# After swapping
print("\nAfter swapping:")
print("x =", x)
print("y =", y)
```

**Explanation**:

- **Step 1**: We declare and initialize **x** and **y** with integer values.
- **Step 2**: Before swapping, we print the original values of **x** and **y**.
- **Step 3**: To swap the values without using a temporary variable, we utilize arithmetic operations and do these 3 operations: <br>
    **x = x + y**: We add **x** and **y** and store the result in **x**. <br>
    **y = x - y**: We subtract the original value of **y** from the updated **x** and store the result in **y**. At this point, **y** holds the original value of **x**. <br>
    **x = x - y**: We subtract the original value of **x** (now stored in **y**) from the updated **x** to get the original value of **y**, and store the result in **x**. 
- **Step 4**: After swapping, we print the new values of **x** and **y**.
---
- If we would have simply written **x = y** on one line followed by **y = x** on the next line, it would not do the swapping but something else would have happened. Here's what happens step by step: <br>
    **x = y**: This line assigns the value of **y** to **x**. So, **x** now becomes **10**. <br>
    **y = x**: This line assigns the value of **x** to **y**. Since **x** has been assigned the value of **y** in the previous line, this effectively assigns **y** the value of **10**. <br>
    So, after executing these two lines, both **x** and **y** will have the value **10**. <br>
    In other words, we've effectively not swapped the values; instead, both variables end up with the value of the variable on the right side of the assignment. This is because the assignment on the second line **y = x** is executed after **x** has already been assigned the value of **y**, leading to both variables having the same value.
---
- There's another way we can do the swapping without using any temporary variable and that too in a single line. <br>
    **x, y = y, x**: This single statement uses tuple unpacking to simultaneously swap the values of **x** and **y** without using a temporary variable. Whenever we want to swap the values of the variables, this is the most efficient and preferred way.
    
    Here's the program rewritten with this one-line swapping technique:

```python
# Declare and initialize variables
x = 50
y = 100

# Before swapping
print("Before swapping:")
print("x =", x)
print("y =", y)

# One-line swap without using a temporary variable
x, y = y, x

# After swapping
print("\nAfter swapping:")
print("x =", x)
print("y =", y)
```

**Ques 2.** Create a program that calculates the area of a rectangle. Take the length and width as inputs from the user and store them in variables. Calculate and display the area.

**Ans 2.**

```python
# Display an introduction of the program to the user
print("Enter length & width of a rectangle and this program will give it's area!")

# Prompt the user to input the length and width of the rectangle
length = float(input("Enter the length of rectangle (in cm): "))
width = float(input("Enter the width of rectangle (in cm): "))

# Calculate the area (length * width) and display it
print(f"Area of the rectangle is {length * width} sq.cm")
```

**Explanation:**
- **Step 1**: We prompt the user to input the length and width of the rectangle using the **input()** function & we convert the input values to a **float**.
    - In the above code, we have done **Explicit Type Conversion (Casting)** for the inputs of length and width variables.
    - This is because the **input()** function, which is used to take input from the user, takes a **string** input by default.
    - If we don't do this type conversion then we'll not be able to multiply length and width, as string data type doesn't support multiplication (`*`) operation.
    - Moreover, as length and width of a rectangle can be any decimal numbers, we must use **float()** along with **input()**.
    - The constructor function **float()** constructs a float number from an integer literal, a float literal or a string literal (provided the string represents a float or an integer). 
- **Step 2**: We calculate the area of the rectangle by multiplying the length and width & we display the calculated area using the print function.
    - In the **print()** statement, we have used the **f-string formatting**, so that the expression **length * width** (which is formula for area) can be embedded inside the string using curly braces.
    - The f-string formatting provides a concise and convenient way to embed expressions inside string literals.

**Ques 3.** Write a Python program that converts temperatures from Celsius to Fahrenheit. Take the temperature in Celsius as input, store it in a variable, convert it to Fahrenheit, and display the result.

**Ans 3.**

```python
# Display an introduction of the program to the user
print("Please input any temperature in Celsius & this program will give you equivalent Fahrenheit temperature!!!")

# Prompt the user to input the temperature in Celsius
t = float(input("What's the temperature in Celsius? -- "))

# Convert Celsius to Fahrenheit
f = (t * 9/5) + 32

# Display the result
print("{} Celsius is equivalent to {} Fahrenheit".format(t,f))
```

**Explanation:**

- **Step 1**: We prompt the user to input the temperature in Celsius using the input function & we convert the input value to float to handle decimal inputs. The input must be typecasted to a float type for the same reason as in question 2.
- **Step 2**: We convert the temperature from Celsius to Fahrenheit using the formula **' _f = (t * 9/5) + 32_ '**.
- **Step 3**: Finally, we display both the Celsius temperature and the equivalent Fahrenheit temperature using the print function.
    - This time in the **print()** statement we have used the **str.format()** method of strings.
    - The curly braces inside the string literals are called as placeholders which will get replaced in order by values given with format() method.
    - First pair of curly braces are replaced with the first value that is **t**.
    - Second pair of curly braces are replaced with the second value that is **f**.
    - Using this **str.format()** method we can handle complex string formatting more efficiently.

**Ques 2.** Create a program that takes a sentence from the user and counts the number of vowels (a,e,i,o,u) in the string.

**Ans 2.**

```python
# Display an introduction of the program to the user
print("Let's count the number of vowels in a sentence you type!")

# Prompt the user to input a sentence
sentence = input("Please type a sentence: ")

# Define a variable to store the count of vowels and set it's value to zero
vowel_count = 0

# Iterate through each character in the sentence
for char in sentence:
    # Check if the character is a vowel (a, e, i, o, u)
    if char.lower() in 'aeiou':
        # If the character is a vowel, increment the vowel count
        vowel_count += 1
        
# Display the count of vowels in the sentence
print("Number of vowels in the sentence: ",vowel_count)
```

**Explanation:**
1. We prompt the user to input a sentence using the **input()** function.

2. We define a variable **vowel_count** to store the count of vowels, initially set to **0**.

3. We iterate through each character in the sentence using a **for** loop.

    **for** loop is used for iterating over any sequence (that is either a list, a tuple, a dictionary, a set, or a string).
    
    Note that we have to use indentation to make sure that we're inside the **for** loop.
        
4. Within the loop, we check if the current character is a vowel. If the character is a vowel, we increment the **vowel_count** variable.

    To handle both uppercase and lowercase vowels, we've first converted the character to equivalent lowercase character using the **lower()** method for strings in Python, and then compared the character with the string **'aeiou'**. Alternatively, we could have written the **if** statement as below (without converting the characters to lowercase and using the string **'AEIOUaeiou'** of both uppercase vowels and lowercase vowels):
    
    **if char in 'AEIOUaeiou':**
    
    The **if** statement checks a condition and when this conditon is **True** the indented statements below the **if** statement are executed. Here, the only indented statement below the **if** statement is to increase the value of the variable **vowel_count** by 1 every time the stated condtion is **True**.
    
    Note that the **in** operator plays an important role in the condtion of the **if** statement. The **in** operator determines whether a given value is a constituent element of a sequence such as a string, list, or tuple. When used in a condition, the statement returns a Boolean result of **True** or **False**. The statement returns **True** if the specified value is found within the sequence, and returns **False** otherwise.
    
5. After iterating through all characters, we display the count of vowels in the sentence using the **print()** function.

**Ques 3.** Given a string, reverse the order of characters using string slicing and print the reversed string.

**Ans 3.**

```python
# Display introduction message to the user
print("Please type a string and you will get to see it reversed!")

# Get the input string from the user
input_string = input()

# Reverse the order of characters using string slicing
reversed_string = input_string[::-1]

# Display the reversed string along with appropriate message
print("Here's the reversed string!")
print(reversed_string)
```

**Explantion:**

- **Step 1**: We prompt the user to input a string using the **input()** function.
- **Step 2**: We reverse the order of characters in the input string using string slicing and we store the reversed string in the variable **reversed_string**.
    - To understand string slicing we must the know about the **indexing** in a string.
    - In a string, when indexing statrs from the left-side it is called as **postive indexing**. In this case, the indexing is **0 1 2 3 4 5 ...**
    - On the other hand, when indexing starts from the right-side it is called as **negative indexing**. In this case, the indexing is **... -5 -4 -3 -2 -1**.
    - The syntax for slicing a string is: **str[start:end:step]**
    - As **start** is not specified so by default start index is considered as the beginning of the string.
    - Also **end** is not specified so by default end index is considered as the end of the string.
    - Lastly, **step** is specified as -1 and this means that the output will be displayed by moving one step backward through the specified slice of the string. 
    - As a result **[ : : -1]** can be used to effectively reverse the string.
- **Step 3**: Finally, we display the reversed string using the **print()** function.

**Ques 4.** Write a program that takes a string as input and checks if it is a palindrome (reads the same forwards and backwards).

**Ans 4.**

```python
# Prompt the user to input a string
print("Please type a string and the program will check if it is a palindrome!")
input_string = input()

# Remove spaces and convert the input string to lowercase for case-insensitive comparison
input_string = input_string.replace(" ", "").lower()

# Check if the input string is equal to its reverse
if input_string == input_string[::-1]:
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

**Explanation:**
- **Step 1**: We prompt the user to input a string using the **input()** function.

- **Step 2**: We remove spaces from the input string using the **replace()** method and convert it to lowercase using the **lower()** method. This step is done to make the comparison case-insensitive and to consider strings with spaces as palindromes.

- **Step 3**: We then compare the original string with its reverse (**input_string[::-1]**) using the equality operator (**==**).

    - If the input string is equal to its reverse, we print that the string is a palindrome; otherwise, we print that it is not a palindrome. For that, **if-else** statement is used.
    
    - The syntax for **if-else** is as below: <br>
    
    **if** Condition:
    
        block of code if conditon is True
        
    **else**:
    
        block of code if condition is False
        
    - Note that in the condition of **if** statement we make use of the **==** sign which is a 'Comparision Operator' that returns **True** if both the operands (left-side value and right-side value) are equal to each other, and returns **False** otherwise.

**Ques 5.** Cretae a program that takes a string as input and removes all the spaces from it. Print the modified string without spaces.

**Ans 5.**

**Method 1: using `replace()` function**

```python
# Prompt the user and get the input string from the user
print("Give a string and this program will remove all the whitespaces from it!")
input_string = input()

# Remove spaces using replace() function
# We replace all the " " substrings i.e spaces with "" substrings i.e empty substrings
# The variable 'modified_string' stores the modified string 
modified_string = input_string.replace(" ", "")

# Print the modified string without spaces
print("Below is the string with all the whitespaces removed:\t",modified_string)
```

**Explanation:**
- In the above code, the **replace()** function is used whose syntax is: **string.replace(oldvalue, newvalue, count)**
    - It is mandatory to specify the **oldvalue** that is the substring which is to be searched for and replaced.
    - It is also mandatory to specify the **newvalue** that is the substring which is going to replace the **oldvalue**.
    - It is not mandatory to specify the **count**. If nothing is specified then all occurrences of the **oldvalue** will be replaced, and if a **count** is specified then that many occurrences of the **oldvalue** will be replaced by the **newvalue**.

- One thing to note in the above code is the **\t** used in the last **print** statement.
    - In Python strings, the backslash (`\`) is a special character, also called the "escape" character. It is used in representing certain whitespace characters.
    - **\t** means add or insert a tab in between.

**Method 2: using `split()` and `join()` functions**

```python
# Prompt the user and get the input string from the user
print("Give a string and this program will remove all the whitespaces from it!")
input_string = input()

# Remove spaces using split() and join()
modified_string = ''.join(input_string.split(" "))

# Print the modified string without spaces
print("Below is the string with all the whitespaces removed: \n",modified_string)
```

**Explanation:**
- In the above code, **split()** and **join()** functions are used.
- The **split()** function splits a string into a list. 
- The syntax for **split()** function is: **string.split(separator, maxsplit)**
    - It is not mandatory to specify the **separator**. By default separator is any whitespace, however you can specify the separator.
    - It is also not mandatory to specify the **maxsplit**. By default at all the occurrences of the separator a split is created, however if specified then it works like this: for example a maxsplit value of 2 will return a list with 3 elements.
- Next function that we've to undertand is the **join()** function.
- The syntax for **join()** function is: **string.join(iterable)**
    - The **join()** function takes all items in an iterable and joins them into one string.
    - It is mandatory here to give an iterable as a paramter inside the join function. Also, a string must be specified which will be the separator in between the elements of the iterable joined into a single string as output.
- Last one thing to note in the above code is the **\n** used in the last **print** statement. When inserted in a string all the characters after **\n** are added to a new line.

**Method 3: using `for` loop and `if` conditional statement**

```python
# Prompt the user and get the input string from the user
print("Give a string and this program will remove all the whitespaces from it!")
input_string = input()

# Initialize an empty string to store the modified string without spaces
modified_string = ''

# Iterate through each character in the input string
for char in input_string:
    # Check if the current character is not a space
    if char != ' ':
        # If not a space, concatenate it to the modified string
        modified_string += char

# Print the modified string without spaces
print("Below is the string with all the whitespaces removed:")
print(modified_string)
```

**Explanation:**
- In the above code, all the things that were used have been explained in the previous questions except for the **!=** sign which is a 'Comparision Operator' named **'Not Equal To'**. It returns **True** if operands on either side are not equal to each other, and returns **False** if they are equal.
