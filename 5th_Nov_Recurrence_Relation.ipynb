### Ques 1: Find the value of `T(2)` for the recurrence relation `T(n)=3T(n-1)+12n`, given that `T(0)=5`.

### Solution:
To find the value of $ T(2) $ for the given recurrence relation $ T(n) = 3T(n−1) + 12n $ with the initial condition $ T(0)=5 $, we can use the recurrence relation to recursively calculate the values. <br><br>
Let's calculate $ T(1) $ first: <br><br>
$ T(1) = 3T(0) + 12×1 = 3×5 + 12 = 15 + 12 = 27 $ <br><br>
Now, let's calculate $ T(2) $ using the same recurrence relation: <br><br>
$ T(2) = 3T(1) + 12×2 = 3×27 + 24 = 81 + 24 = 105 $

### Ques 2: Given a recurrence relation, solve it using the substitution method: <br><br> `a. T(n) = T(n-1)+c` <br><br> `b. T(n) = 2T(n/2)+n` <br><br> `c. T(n) = 2T(n/2)+c` <br><br> `d. T(n) = T(n/2)+c`

### Solution: 2.a.
The given recurrence relation is <br>
$ T(n) = T(n-1) + c $ ..............(1) <br>

So, we can write <br>
$ T(n-1) = T(n-2) + c $ <br>
$ T(n-2) = T(n-3) + c $ <br>
$ T(n-3) = T(n-4) + c $ and so on. <br>

Substituting these values in equation $ (1) $ gives <br>
$ T(n) $ <br>
$ = T(n-1) + c $ <br>
$ = [T(n-2) + c] + c $ <br>
$ = T(n-2) + 2c $ ..............(2) <br>
$ = [T(n-3) + c] + 2c $ <br>
$ = T(n-3) + 3c $ ..............(3) <br>
$ = [T(n-4) + c] + 3c $ <br>
$ = T(n-4) + 4c $ ..............(4) <br>

Observing the pattern, in general we can write <br>
$ T(n) = T(n-i) + ic $ <br>
where $ i=1,2,3,4,...$,value at which base case is reached <br>

Let us assume that the base case is $ T(0)=a $ (representing constant time). <br>

Also, suppose we reach to the base case at $ k^{th} $ step. <br>

In other words, we have $ i=1,2,3,4,...,k $. <br>

At the base case, $ n-k=0  ⇒  k=n $. <br>

So, we have $ i=1,2,3,4,...,n $. <br>

And the recurrence relation simplifies as below: <br>
$ T(n) $ <br>
$ = T(n-k) + kc $ <br>
$ = T(0) + nc $ <br>
$ = a + nc $ <br>
$ = O(n) $ <br>
Hence, the time complexity of the algorithm described by the given recurrence relation is $ O(n) $. <br><br>

### Solution: 2.b.
The given recurrence relation is <br>
$ T(n) = 2T(\frac{n}{2}) + n  ..............(1) $ <br>

So, we can write <br>
$ T(\frac{n}{2}) = 2T(\frac{n}{2^{2}}) + \frac{n}{2} $ <br><br>
$ T(\frac{n}{2^{2}}) = 2T(\frac{n}{2^{3}}) + \frac{n}{2^{2}} $ <br><br>
$ T(\frac{n}{2^{3}}) = 2T(\frac{n}{2^{4}}) + \frac{n}{2^{3}} $ and so on. <br>

Substituting these values in equation $ (1) $ gives <br>
$ T(n) = 2T(\frac{n}{2}) + n $ <br><br>
$ = 2 [2T(\frac{n}{2^{2}}) + \frac{n}{2}] + n $ <br><br>
$ = 2^{2} T(\frac{n}{2^{2}}) + 2n  ..............(2) $ <br><br>
$ = 2^{2} [2T(\frac{n}{2^{3}}) + \frac{n}{2^{2}}] + 2n $ <br><br>
$ = 2^{3} T(\frac{n}{2^{3}}) + 3n  ..............(3) $ <br><br>
$ = 2^{3} [2T(\frac{n}{2^{4}}) + \frac{n}{2^{3}}] + 3n $ <br><br>
$ = 2^{4} T(\frac{n}{2^{4}}) + 4n  ..............(4) $ <br><br>
     
Observing the pattern, in general we can write <br>
$ T(n) = 2^{i} T(\frac{n}{2^{i}}) + in $ <br>
where $ i=1,2,3,4,...,$ value at which base case is reached <br>

Let us assume that the base case is $ T(1)=a $ (representing constant time). <br>

Also, suppose we reach to the base case at $ k^{th} $ step. <br>

In other words, we have $ i=1,2,3,4,...,k $. <br>

At the base case, $ \frac{n}{2^{k}}=1  ⇒ 2^{k}=n  ⇒  k=log_{2}n $. <br>

So, we have $i=1,2,3,4,...,log_{2}n$. <br>

And the recurrence relation simplifies as below: <br>
$ T(n) $ <br>
$ = 2^{k} T(\frac{n}{2^{k}}) + kn $ <br><br>
$ = 2^{log_{2}n} T(1) + (log_{2}n)n $ <br><br>
$ = n^{log_{2}2} a + n log_{2}n $ <br><br>
$ = na + n log_{2}n $ <br><br>
$ = O(n log_{2}n) $ <br><br>
Hence, the time complexity of the algorithm described by the given recurrence relation is $ O(n log_{2}n) $. <br><br>

### Ques 3: Given a recurrence relation, solve it using the recursive tree approach: <br><br> `a. T(n) = 2T(n-1)+1` <br> <br> `b. T(n) = 2T(n/2)+n`

**Steps to solve a recurrence relation using the recursive tree approach:**

**Step 1:** Drawing the recursion tree (i.e. starting with the root/initial term of the recurrence relation and breaking down the problem into subproblems, simulataneously considering the cost incurred at each node).

**Step 2:** Determining the height of the recursion tree (i.e. the number of levels in the tree).

**Step 3:** Calculating the cost at each level (i.e. the time required to solve all the subproblems at that level).

**Step 4:** Determining the cost for the entire tree (i.e. summation of the costs at each level, simplifying the summation to obtain a closed form expression for the total cost, and finally expressing that closed form in big-O notation to determine the time complexity of the algorithm).

### Solution: 3.a.

**Step 1: Drawing the recursion tree:** <br>
The given recurrence relation is <br>
$ T(n) = 2T(n-1) + 1 $ <br>
$ ⇒ T(n) = T(n-1) + T(n-1) + 1 $ <br>
* Therefore, to draw the recursion tree, we start with the root as $ n $ (the size of the problem), and break it into two nodes representing two subproblems of size $ n-1 $. Simulatenously, we write down the cost incurred at the root node which is equivalent to 1.
* This process is carried one step further by expanding each node of size $ n-1 $ into two nodes representing the subproblems of size $ n-2 $ and simulataneously considering the cost incurred at each node as 1.
* We continue expanding each node in the tree by breaking it into its constituent parts as determined by the given recurrence relation.

**Step 2: Determining the height of the recursion tree:** <br>

* The subproblem size for a node at depth $ k $ is $ n-k $. As we descend the tree from the root, the subproblem size hits the base case $ T(0) $ when $ n-k=0 $ or, equivalently, when $ k=n $. Thus, the tree has internal nodes at depths $ 0,1,2,3,...,n-1 $ and leaves at depth $ n $.

**Step 3: Determining the cost at each level of the recursion tree:** <br>

* By observing, we can detect a pattern that is being followed at each level.
* Each level has two times as many nodes as the level above, and so the number of nodes at depth $ i $ is $ 2^{i} $. Each internal node at depth $ i=0,1,2,3,...,n-1,n $ has a cost of 1. Multiplying, we see that the total cost of all nodes at a given depth i is $ 2^{i}*1=2^{i} $.
* At level 0, there are $ 2^{0} $ nodes so the cost of this level is $ 2^{0} $.
* At level 1, there are $ 2^{1} $ nodes so the cost of this level is $ 2^{1} $.
* At level 2, there are $ 2^{2} $ nodes so the cost of this level is $ 2^{2} $ and so on till the bottom level.

**Step 4: Determining the cost for the entire tree:** <br>

Now we add up the cost over all levels to determine the cost for the entire tree. <br>
$ T(n) = 2^{0} + 2^{1} + 2^{2} + 2^{3} + ... + 2^{n-1} + 2^{n} $ <br>
$ = \frac{2^{n+1}-1}{2-1} $ (by using the formula for the sum of finite terms of a Geometric Progression) <br><br>
$ = {2^{n+1}-1} $ <br>
$ = {2*2^{n}-1} $ <br>
$ = O(2^{n}) $ <br>
Hence, the time complexity of the algorithm described by the given recurrence relation is $ O(2^{n}) $. <br><br>

### Solution: 3.b.

The given recurrence relation is <br>
$ T(n) = 2T(\frac{n}{2}) + n $ <br>
$ ⇒ T(n) = T(\frac{n}{2}) + T(\frac{n}{2}) + n $ <br>

* Therefore, to draw the recursion tree, we start with the root as $ n $ (the size of the problem), and break it into two nodes representing two subproblems of size $ \frac{n}{2} $. Simulatenously, we write down the cost incurred at the root node which is equivalent to $ n $.
* This process is carried one step further by expanding each node of size $ \frac{n}{2} $ into two nodes representing the subproblems of size $ \frac{n}{2^{2}} $ and simulataneously considering the cost incurred at each node as $ \frac{n}{2} $.
* We continue expanding each node in the tree by breaking it into its constituent parts as determined by the given recurrence relation.

**Step 2: Determining the height of the recursion tree:** <br>

* The subproblem size for a node at depth $ k $ is $ \frac{n}{2^{k}} $. As we descend the tree from the root, the subproblem size hits base case $ T(1) $ when $ \frac{n}{2^{k}}=1 $ or, equivalently, when $ n=2^{k} $. Solving for the value of $ k $ gives $ k=log_2{n} $. Thus, the tree has internal nodes at depths $ 0,1,2,3,...,log_2{(n-1)} $ and leaves at depth $ log_2{n} $.

**Step 3: Determining the cost at each level of the recursion tree:** <br>

* Each level has two times as many nodes as the level above, and so the number of nodes at depth $ i $ is $ 2^{i} $. Each internal node at depth $ i=0,1,2,3,...,n-1,n $ has a cost of $ \frac{n}{2^{i}} $ . Multiplying, we see that the total cost of all nodes at a given depth $ i $ is $ 2^{i}*\frac{n}{2^{i}}=n $.

**Step 4: Determining the cost for the entire tree:** <br>

Now we add up the cost over all levels to determine the cost for the entire tree. <br>
$ T(n) = n + n + n + n + ... + n + n $ ($ k $ times) <br>
$ = k*n $ <br>
$ = nlog_2{n} $ <br>
$ = O(nlog_2{n}) $ <br>
Hence, the time complexity of the algorithm described by the given recurrence relation is $ O(nlog_2{n}) $.
